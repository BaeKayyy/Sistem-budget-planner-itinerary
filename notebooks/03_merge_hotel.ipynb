```python
import pandas as pd

df1 = pd.read_csv("../data/raw/hotel.csv")
df2 = pd.read_csv("../data/raw/hotel2.csv")
df3 = pd.read_csv("../data/raw/hotel3.csv")

df_hotel = pd.concat([df1, df2, df3], ignore_index=True)

# hapus duplikat
df_hotel = df_hotel.drop_duplicates(subset=["name"])

# tambah tipe
df_hotel["type"] = "hotel"

df_hotel.to_csv("../data/processed/hotel_final.csv", index=False)

print("Total hotel:", len(df_hotel))
```
