```python
import pandas as pd

df1 = pd.read_csv("../data/raw/kuliner1.csv")
df2 = pd.read_csv("../data/raw/kuliner2.csv")

df_kuliner = pd.concat([df1, df2], ignore_index=True)

# hapus duplikat
df_kuliner = df_kuliner.drop_duplicates(subset=["name"])

# tambah tipe (penting buat nanti)
df_kuliner["type"] = "kuliner"

df_kuliner.to_csv("../data/processed/kuliner_final.csv", index=False)

print("Total kuliner:", len(df_kuliner))
```
