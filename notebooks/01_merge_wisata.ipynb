```python
import pandas as pd

df1 = pd.read_csv("../data/raw/wisata jogja.csv")
df2 = pd.read_csv("../data/raw/wisata2.csv")

df_wisata = pd.concat([df1, df2], ignore_index=True)

# hapus duplikat
df_wisata = df_wisata.drop_duplicates(subset=["name"])

df_wisata.to_csv("../data/processed/wisata_final.csv", index=False)

print("Total wisata:", len(df_wisata))
```
